```python
def dimacs_read(dimacs):
    variables = dict()
    with open(dimacs, 'r') as f:
        dimacs_lines = f.readlines()
        for line in dimacs_lines:
            if line.startswith("c"):
                m = re.search('(c) (\d+) (\w+(=[.\w+/-]+)?)', line)
                var_id = int(m.group(2))
                var_name = m.group(3)  
                variables[var_id] = var_name
    return variables
```

```python
d = dimacs_read("examples/out.dimacs")
```
